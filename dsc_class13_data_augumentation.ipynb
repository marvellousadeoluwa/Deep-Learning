```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
%matplotlib inline
```

```
 from tensorflow.keras.preprocessing.image import img_to_array, load_img, array_to_img, ImageDataGenerator
```

```
# Load img and then convert array
img = load_img("IMG-20201202-WA0076.jpg")
img_array = img_to_array(img)
img_array.shape
```

```
#np.expand_dims just helps you expand the dimension of the image you have to include the number of images you have at hand.
#for instance one image size can be (224,224,3)
#using expand dims you get(1,224,224,3)
 
samples = np.expand_dims(img_array,0)
samples.shape
```

# Horizontal Shift Augumentation

```
# Create img data augumentation generator
datagen = ImageDataGenerator(width_shift_range= [-200,200])
# create your iterator
it = datagen.flow(samples, batch_size=1)
```

```
for i in range(9):
  plt.subplot(330+1+i)
  batch = it.next()
  image = batch[0].astype("uint8")
  plt.imshow(image)
```
